```python
#Import dependencies 
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = pd.read_csv("D:\python programs\dspl\\archive\ADANIPORTS.csv") 
df2 = df
```

```python
#Checking outliers using boxplot
print(plt.figure(figsize = (10, 4), dpi = 100)) 
sns.boxplot(x = "Open", data = df2) 
```

```python
#Visualizing by Highs of adaniports by countplot
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("Accent_r") 
sns.set_palette(color_palette)
sns.countplot(x = "High", data = df2) 
```

```python
#Count of closing of adaniports
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("cool")
 
sns.set_palette(color_palette) 
sns.countplot(x = "Close", data = df2) 
```

```python
# # #Distribution of CGPA plt.figure(figsize = (10, 4), dpi = 100) 
grp = dict(df2.groupby('Symbol').groups) 
m = {}
for key, val in grp.items(): 
    if key in m:
        m[key] += len(val) 
    else:
        m[key] = len(val) 
        plt.title("percentage records till date") 
        plt.pie(m.values(), labels = m.keys())
```

```python
# open analysis as per symbol
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("magma") 
sns.set_palette(color_palette)
sns.barplot(x = "Symbol", y = "Open", data = df2) 
```

```python
# open analysis as per date
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("plasma") 
sns.set_palette(color_palette)
 
sns.barplot(x = "Date", y = "Open", data = df2) 
```

```python
# Close as per symbol
plt.figure(figsize = (10, 4), dpi = 100)
color_palette = sns.color_palette("plasma") 
sns.set_palette(color_palette)
sns.barplot(x = "Symbol", y = "Close", data = df2) 
```

```python
# Finding correlation 
plt.figure(figsize = (10, 6), dpi = 100) 
color = sns.color_palette("magma")
sns.heatmap(df2.corr(), vmax=0.9, annot=True,cmap = color)
```
